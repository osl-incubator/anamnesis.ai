```python
import os
import json
from pathlib import Path

from dotenv import load_dotenv
from anamnesisai.core import AnamnesisAI

NOTEBOOK_DIR = Path(os.getcwd())
ENV_FILE = (NOTEBOOK_DIR.parent.parent / "tests" / ".env").resolve()

load_dotenv(ENV_FILE)

OPENAI_API_KEY = os.environ["OPENAI_API_KEY"]
```

```python
# initialize AnamnesisAI

aai = AnamnesisAI(backend="openai", api_key=OPENAI_API_KEY)
```

```python
#using anaminesisai to check possible fhir resource present in the dialog

fhir_data = aai._check_possible_fhir_resources(CAR0001)

print(fhir_data)
```

```python
# check number of resources detected (True)
# snippet from core.py
# count n of resources found
from anamnesisai.supported_fhir import (
    RESOURCES_CLASSES,
    FHIRResourceFoundModel,
)

n_of_resources = 0
for fhir_class in RESOURCES_CLASSES:
    resource_name = fhir_class.__name__

    if not getattr(fhir_data, resource_name, False):
        continue
    else:
        n_of_resources += 1

print("number of resources:", n_of_resources)
```

```python
# extract fhir resources

fhir_resources, invalid_fhir_resources = aai.extract_fhir(CAR0001)
```

```python

print(f"fhir_resources: {type(fhir_resources)}<Resource>, invalid_fhir_resources: {type(invalid_fhir_resources)}<InvalidFHIRResource> ")
```

```python
# succesful returned resources
print(f"Number of fhirs resources: {len(fhir_resources)}")
print(f"Number of invalid  resources: {len(invalid_fhir_resources)}")
```

```python
print("Valid FHIR RESOURCES:")
print(json.dumps([json.loads(fhir_resource.model_dump_json()) for fhir_resource in fhir_resources], indent=4))
```

```python
print("Invalid FHIR RESOURCES:")
print(json.dumps([json.loads(fhir_resource.model_dump_json()) for fhir_resource in invalid_fhir_resources], indent=4))
```
